```python
from prophet import Prophet
```

```python
df = df.rename(columns={"–î–∞—Ç–∞": "ds", "–ü—Ä–æ—Å–º–æ—Ç—Ä–æ–≤": "y"})
ts = df[['ds', 'y']]
ts.ds = ts.ds.apply(lambda x: x[:10])
ts.y = ts.y.apply(lambda x: int(x))
```

```python
date_dict = {}
for index, row in ts.iterrows():
    if row['ds'] in date_dict:
        date_dict[row['ds']] += row['y']
    else:
        date_dict[row['ds']] = row['y']
```

```python
ts = pd.DataFrame(date_dict.items(), columns=['ds', 'y'])
ts = ts.sort_values('ds')
ts.ds = ts.ds.apply(lambda x: pd.to_datetime(x[:10], format='%Y-%m-%d'))
train = ts[ts.ds < pd.to_datetime("2021-12-01", format='%Y-%m-%d')]
test = ts[ts.ds >= pd.to_datetime("2021-12-01", format='%Y-%m-%d')]
```

```python
ax = ts.plot(x="ds", y="y", label="real")
```

```python
train.sort_values('ds')
```

```python
test
```

```python
model = Prophet()
model.fit(train)
preds = model.predict(test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error

print(f"MAPE: {round(mean_absolute_percentage_error(preds.yhat, test.y)*100, 2)} %")
```

```python
ax = preds.plot(x="ds", y="yhat", label="predictions")
test.plot(x="ds", y="y", ax=ax, label="real")
plt.show()
```
